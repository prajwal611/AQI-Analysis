```python
df["number_of_monitoring_stations"]=df["number_of_monitoring_stations"].fillna(0)
```

```python
df["date"]=pd.to_datetime(df["date"])
df["aqi_value"]=pd.to_numeric(df["aqi_value"])
```

```python
import sqlite3
conn=sqlite3.connect("aqi_database.db")
df.to_sql("aqi_data",conn,if_exists="replace",index=False)
```

```python
query="""
Select state,AVG(aqi_value) AS avg_aqi from aqi_data group by state order by avg_aqi DESC;
"""
result=pd.read_sql_query(query,conn)
result.head()
```

```python
query2="""
select area,MAX(aqi_value) as max_aqi from aqi_data group by area order by max_aqi desc limit 10;
"""
result2=pd.read_sql_query(query2,conn)
result2.head()
```

```python
query3="""
select air_quality_status,count(*) as total from aqi_data
group by air_quality_status;
"""
result3=pd.read_sql_query(query3,conn)
result3.head()
```

```python
query4="""
select prominent_pollutants,count(*) as occurences from aqi_data group by prominent_pollutants order by occurences desc;
"""
result4=pd.read_sql_query(query4,conn)
result4.head()
```

```python
query5="""
select state, SUM(number_of_monitoring_stations) from aqi_data group by state order by sum(number_of_monitoring_stations) desc;
"""
result5=pd.read_sql_query(query5,conn)
result5.head()
```
